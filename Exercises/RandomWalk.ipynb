```python
import numpy as np
import numpy.random
import matplotlib.pyplot as plt

%matplotlib inline

# Seed for X random process
seedX = 1010

# Seed for Y random process
seedY = 1110

# X random process generator
rpgX = np.random.RandomState(seedX)

# Y random process generator
rpgY = np.random.RandomState(seedY)
```

```python
# Number of points to be generated
NX = 10000

arrayX = rpgX.rand(NX)

#print("ArrayX", arrayX)

#print("ArrayX", arrayX.reshape(2,5,order='F'))

# number of lines
nl = np.int(NX/2)

arrayX2 = arrayX.reshape(2,nl,order='F')

# print(np.shape(arrayX),np.shape(arrayX2))

plt.xlim((0.0, 1.0))
plt.ylim((0.0, 1.0))
#plt.axis('equal')
plt.plot(arrayX2[0,:],arrayX2[1,:],'.')
```

```python
# Number of points to be generated
N = 10000

# number of lines
nl = np.int(N/2)

arrayX = rpgX.rand(nl)
arrayY = rpgY.rand(nl)

plt.xlim((0.0, 1.0))
plt.ylim((0.0, 1.0))

plt.plot(arrayX,arrayY,'.')
```
